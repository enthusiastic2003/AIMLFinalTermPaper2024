```python
acc_list_class = []

for i in range(10):
    acc_list_class.append(pd.read_csv(f'SelectiveForgetting/data/acc{i}_{alpha}.csv', index_col=0))
    
```

```python
accs_class = pd.concat(acc_list_class)
```

```python
accs_class = accs_class.reset_index(drop=True)
column_names = ['Original', 'Fine-tuned', 'Fisher', 'Re-trained']
accs_class.columns = column_names
accs_class = accs_class * 100
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=accs_class, ax=ax)
ax.plot(accs_class.index, accs_class['Fisher'], 'o', color='green', markersize=4)
ax.plot(accs_class.index, accs_class['Fine-tuned'], 'o', color='orange', markersize=4)
ax.plot(accs_class.index, accs_class['Original'], 'o', color='royalblue', markersize=4)
ax.plot(accs_class.index, accs_class['Re-trained'], 'o', color='red', markersize=4)
ax.set_xlabel('Class')
ax.set_ylabel('Accuracy (%)')
ax.set_title('Accuracy on Retained Data for MNIST')
ax.set_xticks([i for i in range(10)])
plt.savefig('accc.png')
```
